# Searching the pontential Desired Patients for each Experiment:

At the end of these file we will download 2 dataframes contaning in each row a record of each diagnosis. One dataframe contains the cases after a given year, and the other one before that year. For each diagnosis we will encode the patiets unique identifier (at column eid), the code of the given subtype of the disease (at column diagnosis) and the date of the disorder (at column date). We can use such files for future conmorbidity analysis.

```python
import numpy as np
import pandas as pd
```

```python
# File containing all types of dementia codes:
alld_dates = pd.read_csv("alld_dates.tab",delimiter="\t", header=None)
# File containing all types of dementia diagnosis date:
alld_codes = pd.read_csv("alld_codes.tab",delimiter="\t", header=None)
```

```python
# I sorted both files in increasing order by patients eids with are unique and common and re-set the index for
# convinience. We will end up having both dataframes equally arranged.
alld_codes = alld_codes.sort_values([1], axis=0,ascending=[True])
alld_codes = alld_codes.reset_index() 
del alld_codes["index"]
del alld_codes[0]

alld_dates = alld_dates.sort_values([1], axis=0,ascending=[True])
alld_dates = alld_dates.reset_index() 
del alld_dates["index"]
del alld_dates[0]
```

```python
# List of our target ICD-10 codes: In our case are Dementia codes 
list_codes = ["F000","F001","F002","F009","G300","G301","G308","G309","F010","F011","F012","F013","F018","F019","I673","F020","G310","A810","F021","F022","F023","F024","F028","F051","F106","G311","G318"]
```

```python
# We create a dictionary that will store all those records that have been done after baseline assesment. 
# Hence, those that have no issue to be used for risk prediction.
eid_code_date_after = {"eid":[], "code":[], "date":[]}
```

```python
# We create a dictionary that will store all those records that have been done before baseline assesment. 
eid_code_date_before = {"eid":[], "code":[], "date":[]}
```

```python
# Traverse the matrix and
for i in range(0, len(alld_codes.index)):
    for j in range(0, len(alld_codes.columns)):
        # Where it finds a target diagnosis stops to store the date
        if (str(alld_codes.values[i,j]) in list_codes):
            # Create a dictionary with for each diagnosis of each patient and its date that can be used for risk prediction
            if int(str(alld_dates.values[i,j])[0:4])>=2011:
                eid_code_date_after.setdefault("eid", []).append(alld_codes.values[i,0])
                eid_code_date_after.setdefault("code", []).append(alld_codes.values[i,j])
                eid_code_date_after.setdefault("date", []).append(alld_dates.values[i,j])
            # Create a dictionary with for each diagnosis of each patient and its date that can't be used for risk prediction
            else:
                eid_code_date_before.setdefault("eid", []).append(alld_codes.values[i,0])
                eid_code_date_before.setdefault("code", []).append(alld_codes.values[i,j])
                eid_code_date_before.setdefault("date", []).append(alld_dates.values[i,j])
        elif str(alld_codes.values[i,j]) == "nan": break
```

Now for a true potential Dementia screening group of patients, we still miss some steps. We should remove all those eids that also had a diagnosis previous to 2011, hence strored in eid_code_year_before Dataframe. Moreover, we will create a new Dataframe that will only code the eids and 4 bynary columns specifyng which type of Dementia each suffered: AD, VD, FTD, OD. 

```python
AD = ["F000","F001","F002","F009","G300","G301","G308","G309"]
VD = ["F010","F011","F012","F013","F018","F019","I673"]
FTD = ["F020","G310"]
OD = ["A810","F021","F022","F023","F024","F028","F051","F106","G311","G318"]
```

```python
eid_types_after_nobefore = {"eid":[], "AD":[], "VD":[], "FTD":[], "OD":[], "ALLD":[], "Code":[]}
#eid_types_after_nobefore = {"eid":[], "AD":[], "VD":[], "FTD":[], "OD":[], "ALLD":[], "Code":[]}
```

```python
#eliminated_count = len(set(list(eid_code_year_before["eid"])));eliminated_count
```

```python
list_eids_del = list(set(eid_code_year_before["eid"]))
```

```python
list_eids = [] #list that will store all the unique eids
list_types = []  #list that will store what main diagnosis types (AD-VD-FTD-OD) has the current eid
count = 0

for i in range(0, len(eid_code_year_after.index)):
    # We will just keep those patients that haven't had any diagnosis of dementia previous to 2011 
    if int(eid_code_year_after.values[i,0]) not in list_eids_del:
        # If its the first time we explore this eid we will store the information collected from the previous one, and reset the "list_types":
        if str(eid_code_year_after.values[i,0]) not in list_eids:
                      
            list_types = list(set(list_types))
            if i != 0: # we will storage always the previous eid, hence in iteration 0, we cannot enter as it doesn't exists.
                eid_types_after_nobefore.setdefault("eid", []).append(list_eids[-1])
                if int(1) in list_types: eid_types_after_nobefore.setdefault("AD", []).append(int(1))
                else: eid_types_after_nobefore.setdefault("AD", []).append(int(0))
                if int(2) in list_types: eid_types_after_nobefore.setdefault("VD", []).append(int(1))
                else: eid_types_after_nobefore.setdefault("VD", []).append(int(0))
                if int(3) in list_types: eid_types_after_nobefore.setdefault("FTD", []).append(int(1))
                else: eid_types_after_nobefore.setdefault("FTD", []).append(int(0))
                if int(4) in list_types: eid_types_after_nobefore.setdefault("OD", []).append(int(1))
                else:eid_types_after_nobefore.setdefault("OD", []).append(int(0))   
                eid_types_after_nobefore.setdefault("ALLD", []).append(1)
                #eid_types_after_nobefore.setdefault("Code", []).append(str(eid_code_year_after.values[i,1]))

            
            list_types = [] # Every first time that get a different eid, we set the list_types empty and start updating 
            list_eids.append(str(eid_code_year_after.values[i,0]))                       

            if str(eid_code_year_after.values[i,1]) in AD:
                list_types.append(int(1))
            elif str(eid_code_year_after.values[i,1]) in VD:
                list_types.append(int(2))
            elif str(eid_code_year_after.values[i,1]) in FTD:
                list_types.append(int(3))
            elif str(eid_code_year_after.values[i,1]) in OD:
                list_types.append(int(4))
        
        else: # Tis updates the current diagnosis of the patient, hence we will enter to thsi condition when the same patient develops several diagnosis
            if str(eid_code_year_after.values[i,1]) in AD:
                list_types.append(int(1))
            elif str(eid_code_year_after.values[i,1]) in VD:
                list_types.append(int(2))
            elif str(eid_code_year_after.values[i,1]) in FTD:
                list_types.append(int(3))
            elif str(eid_code_year_after.values[i,1]) in OD:
                list_types.append(int(4))            

# As we were updating the dictionary every time we saw a new eid,
# the last one doesn't get recorderd, thats why:
eid_types_after_nobefore.setdefault("eid", []).append(list_eids[-1])
if int(1) in list_types: eid_types_after_nobefore.setdefault("AD", []).append(int(1))
else: eid_types_after_nobefore.setdefault("AD", []).append(int(0))
if int(2) in list_types: eid_types_after_nobefore.setdefault("VD", []).append(int(1))
else: eid_types_after_nobefore.setdefault("VD", []).append(int(0))
if int(3) in list_types: eid_types_after_nobefore.setdefault("FTD", []).append(int(1))
else: eid_types_after_nobefore.setdefault("FTD", []).append(int(0))
if int(4) in list_types: eid_types_after_nobefore.setdefault("OD", []).append(int(1))
else:eid_types_after_nobefore.setdefault("OD", []).append(int(0))
eid_types_after_nobefore.setdefault("ALLD", []).append(1)
#eid_types_after_nobefore.setdefault("Code", []).append(str(eid_code_year_after.values[i,1]))
```
